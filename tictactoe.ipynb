```python
# Función para dibujar el tablero
def dibujar_tablero(tablero):
    print("   A   B   C")
    for i in range(3):
        print(str(i + 1) + " ", end="")
        for j in range(3):
            print(" " + tablero[i][j] + " ", end="")
            if j < 2:
                print("|", end="")
        print()
        if i < 2:
            print("  ---+---+---")

# Función para verificar si un movimiento es válido
def es_movimiento_valido(tablero, fila, columna):
    if fila < 1 or fila > 3 or columna < 1 or columna > 3:
        return False
    if tablero[fila - 1][columna - 1] != " ":
        return False
    return True

def hay_ganador(tablero):
    # Comprobar filas
    for fila in tablero:
        if fila[0] == fila[1] == fila[2] != " ":
            return True

    # Comprobar columnas
    for columna in range(3):
        if tablero[0][columna] == tablero[1][columna] == tablero[2][columna] != " ":
            return True

    # Comprobar diagonales
    if tablero[0][0] == tablero[1][1] == tablero[2][2] != " ":
        return True
    if tablero[0][2] == tablero[1][1] == tablero[2][0] != " ":
        return True

    return False

# Función para comprobar si hay un empate
def hay_empate(tablero):
    for fila in tablero:
        if " " in fila:
            return False
    return True

# Tablero vacío
tablero = [[" ", " ", " "], [" ", " ", " "], [" ", " ", " "]]

# Jugador actual
jugador = "X"

# Variable para controlar el fin del juego
fin_del_juego = False

# Bucle principal del juego
while not fin_del_juego:
    # Dibujar tablero
    dibujar_tablero(tablero)

    # Obtener entrada del jugador
    posicion = input("Jugador " + jugador + ", ingresa la posición (por ejemplo, A1): ")

    # Convertir la posición a fila y columna
    columna = ord(posicion[0].upper()) - ord("A") + 1
    fila = int(posicion[1])

    # Verificar si el movimiento es válido
    if es_movimiento_valido(tablero, fila, columna):
        # Realizar el movimiento
        tablero[fila - 1][columna - 1] = jugador

        # Comprobar si hay un ganador
        if hay_ganador(tablero):
            dibujar_tablero(tablero)
            print("¡Tenemos un ganador! Jugador " + jugador + " ha ganado.")
            fin_del_juego = True

        # Comprobar si hay un empate
        if hay_empate(tablero):
            dibujar_tablero(tablero)
            print("¡Empate! No hay ganador.")
            fin_del_juego = True

        # Cambiar de jugador
        jugador = "O" if jugador == "X" else "X"
    else:
        # Mostrar mensaje de error
        print("Movimiento inválido. Inténtalo de nuevo.")
```
